```python
import os

def createdataframe(dir):
    image_paths = []
    labels = []
    for label in os.listdir(dir):
        for file in os.listdir(os.path.join(dir, label)):
            image_paths.append(os.path.join(dir, label, file))
            labels.append(label)
    return image_paths, labels
```

```python

train = pd.DataFrame()
train['image'], train['label'] = createdataframe(TRAIN_DIR)
train
```

```python

test = pd.DataFrame()
test['image'], test['label'] = createdataframe(TEST_DIR)
test
```

```python
from tqdm.notebook import tqdm
```

```python
from tqdm import tqdm
import numpy as np
from tensorflow.keras.preprocessing.image import load_img

def extrat_features(images):
    features = []
    for image in tqdm(images):   # correct variable
        img = load_img(image, color_mode='grayscale', target_size=(48, 48))
        img = np.array(img)
        features.append(img)
    return np.array(features)
```

```python
color_mode='grayscale'
```

```python
train_features = extrat_features(train['image'])
```

```python
test_features = extrat_features(test['image'])
```

```python
x_train = train_features/255.0
x_test =test_features/255.0
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(train['label'])
```

```python
y_train = le.transform(train['label'])
y_test = le.transform(test['label'])
```

```python
y_train = to_categorical(y_train,num_classes=7)
y_test = to_categorical(y_test,num_classes=7)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense

model = Sequential()

# First block
model.add(Conv2D(128, kernel_size=(3,3), activation='relu', input_shape=(48,48,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

# Second block
model.add(Conv2D(256, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

# Third block
model.add(Conv2D(512, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

# Fourth block
model.add(Conv2D(512, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

# Fully connected layers
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))

# Output layer (7 classes)
model.add(Dense(7, activation='softmax'))

# Optional: compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Load model
model = load_model("emotion_model.h5")

# Emotion labels (must match your dataset folder order)
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

def predict_emotion(img_path):
    # Load image
    img = image.load_img(img_path, color_mode='grayscale', target_size=(48, 48))
    
    # Convert to array
    img_array = image.img_to_array(img)
    img_array = img_array / 255.0
    img_array = np.reshape(img_array, (1, 48, 48, 1))

    # Predict
    prediction = model.predict(img_array)
    emotion = emotion_labels[np.argmax(prediction)]

    return emotion

# Example usage
result = predict_emotion("C:\\Users\\Des\\Desktop\\safalta kaushal\\close-up-portrait-gorgeous-young-woman_273609-40846.avif")
print("Predicted Emotion:", result)
```
